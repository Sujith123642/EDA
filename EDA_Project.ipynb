# Description of the column

FlightID: Unique identifier for each flight.  
Airline: Name of the airline operating the flight.  
FlightNumber: Number assigned to the flight.  
Origin: Departure airport code.  
Destination: Arrival airport code.  
ScheduledDeparture: Planned departure time.  
ActualDeparture: Actual departure time.  
ScheduledArrival: Planned arrival time.  
ActualArrival: Actual arrival time.  
DelayMinutes: Total delay time in minutes.  
DelayReason: Primary reason for the delay.  
Cancelled: Indicates if the flight was canceled (1 for yes, 0 for no).  
Diverted: Indicates if the flight was diverted (1 for yes, 0 for no).  
AircraftType: Type/model of the aircraft used.  
TailNumber: Unique registration number of the aircraft.  
Distance: Distance between the origin and destination airports.

# How many flights are there for each airline?

```python
import pandas as pd
flights_per_airline = df.groupby('Airline').size().reset_index(name='FlightCount')
print(flights_per_airline)


# Plotting with matplotlib
import matplotlib.pyplot as plt
flights_per_airline = df.groupby('Airline').size().reset_index(name='FlightCount')
plt.figure(figsize=(10,6))
plt.bar(flights_per_airline['Airline'], flights_per_airline['FlightCount'], color='skyblue')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.title('Number of Flights for Each Airline')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# What is the average delay time for flights by each airline? 

```python
import pandas as pd
average_delay_per_airline = df.groupby('Airline')['DelayMinutes'].mean().reset_index(name='AverageDelayMinutes')
print(average_delay_per_airline)

# ploting
plt.figure(figsize=(10,6))
plt.bar(average_delay_per_airline['Airline'], average_delay_per_airline['AverageDelayMinutes'], color='orange')
plt.xlabel('Airline')
plt.ylabel('Average Delay (Minutes)')
plt.title('Average Delay Time for Each Airline')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

The plot shows average delay times for each airline, revealing which airlines are more punctual and which face longer delays

# What percentage of flights were cancelled by each airline?

```python
import pandas as pd
# Calculate total flights and cancelled flights per airline
total_flights = df.groupby('Airline').size()
cancelled_flights = df[df['Cancelled'] == 1].groupby('Airline').size()
# Calculate the percentage of cancelled flights
cancelled_percentage = (cancelled_flights / total_flights * 100).reset_index(name='CancelledPercentage')
print(cancelled_percentage)


import pandas as pd
import matplotlib.pyplot as plt
# Calculate total flights and cancelled flights per airline
total_flights = df.groupby('Airline').size()
cancelled_flights = df[df['Cancelled'] == 1].groupby('Airline').size()
# Calculate the percentage of cancelled flights
cancelled_percentage = (cancelled_flights / total_flights * 100).reset_index(name='CancelledPercentage')
# Plotting a pie chart for cancelled flights
plt.figure(figsize=(10, 8))
plt.pie(cancelled_percentage['CancelledPercentage'], labels=cancelled_percentage['Airline'], autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Cancelled Flights by Airline')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

#  What are the most common reasons for flight delays?

```python

import seaborn as sns
sns.set(style="whitegrid")
# Handle Missing Values in 'DelayReason' (Replace NaN with 'Unknown')
df['DelayReason'] = df['DelayReason'].fillna('Unknown')
# Calculate the Frequency of Each Delay Reason
delay_counts = df['DelayReason'].value_counts()
# Plot the Top 10 Most Common Delay Reasons in a Bar Graph
plt.figure(figsize=(12, 8))
sns.barplot(x=delay_counts.values[:10], y=delay_counts.index[:10], palette='viridis')
plt.title('Top 10 Most Common Flight Delay Reasons', fontsize=16)
plt.xlabel('Number of Delays', fontsize=14)
plt.ylabel('Delay Reason', fontsize=14)
plt.tight_layout()
plt.show()
```

# What times of day see the most flight delays?

```python
import pandas as pd
df['ScheduledDeparture'] = pd.to_datetime(df['ScheduledDeparture'])
# Extract the hour of the scheduled departure
df['DepartureHour'] = df['ScheduledDeparture'].dt.hour
# Filter for delayed flights (where DelayMinutes > 0)
delayed_flights = df[df['DelayMinutes'] > 0]
# Group by hour of departure and count the number of delayed flights
delays_by_hour = delayed_flights.groupby('DepartureHour').size()

print(delays_by_hour)
# Plotting the result
plt.figure(figsize=(10,6))
delays_by_hour.plot(kind='bar', color='skyblue')
plt.title('Number of Delays by Hour of Day')
plt.xlabel('Hour of Scheduled Departure')
plt.ylabel('Number of Delayed Flights')
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.show()

```

# Which airlines have the highest rate of flight cancellations?

```python
import pandas as pd
# Filter out canceled flights (assuming 1 = canceled and 0 = not canceled)
cancelled_flights = df[df['Cancelled'] == 1]
# Group by airline and count total flights and canceled flights
total_flights_by_airline = df.groupby('Airline').size()
cancelled_flights_by_airline = cancelled_flights.groupby('Airline').size()
# Calculate the cancellation rate for each airline
cancellation_rate = (cancelled_flights_by_airline / total_flights_by_airline) * 100
# Sort by highest cancellation rate
cancellation_rate = cancellation_rate.sort_values(ascending=False)
print(cancellation_rate)


# Plotting the result
plt.figure(figsize=(10,6))
cancellation_rate.plot(kind='bar', color='salmon')
plt.title('Cancellation Rate by Airline')
plt.xlabel('Airline')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()  # Adjust layout to avoid cut-off labels
plt.show()
```

# How do weather-related delays compare to other delay reasons?



```python
import pandas as pd
# Filter out delayed flights (where DelayMinutes > 0)
delayed_flights = df[df['DelayMinutes'] > 0]
# Group by delay reason and sum the total delay minutes for each reason
delay_by_reason = delayed_flights.groupby('DelayReason')['DelayMinutes'].sum()
# Sort delay reasons by total delay minutes
delay_by_reason = delay_by_reason.sort_values(ascending=False)
# Display the total delay minutes by each reason
print(delay_by_reason)


# Plotting the result
plt.figure(figsize=(10,6))
delay_by_reason.plot(kind='bar', color='lightblue')
plt.title('Total Delay Minutes by Delay Reason')
plt.xlabel('Delay Reason')
plt.ylabel('Total Delay Minutes')
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

#  How do arrival delays correlate with departure delays for the same flights?

```python
import pandas as pd
# Create a new column for arrival delay
df['ArrivalDelay'] = df['ActualArrival'] - df['ScheduledArrival']
# Convert the ArrivalDelay from timedelta to minutes
df['ArrivalDelay'] = df['ArrivalDelay'].dt.total_seconds() / 60
# Filter for relevant flights (where both delays are present)
relevant_flights = df[(df['DelayMinutes'] > 0) & (df['ArrivalDelay'].notnull())]
# Calculate the correlation between departure delays and arrival delays
correlation = relevant_flights['DelayMinutes'].corr(relevant_flights['ArrivalDelay'])
# Display the correlation coefficient
print(f'Correlation between Departure Delays and Arrival Delays: {correlation:.2f}')

plt.figure(figsize=(10, 6))
plt.scatter(relevant_flights['DelayMinutes'], relevant_flights['ArrivalDelay'], alpha=0.5)
plt.title('Correlation between Departure Delays and Arrival Delays')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Arrival Delay (Minutes)')
plt.grid(True)
# Show correlation coefficient on the plot
plt.annotate(f'Correlation: {correlation:.2f}', xy=(0.1, 0.9), xycoords='axes fraction', fontsize=12, color='red')
plt.tight_layout()
plt.show()
```

# What is the distribution of flight distances for delayed flights

```python
# Filter for delayed flights
delayed_flights = df[df['DelayMinutes'] > 0]
# Calculate the average distance for delayed flights
average_distance_delayed = delayed_flights['Distance'].mean()
print(f"\nAverage distance of delayed flights: {average_distance_delayed:.2f} miles")
# Distribution of flight distances for delayed flights
distance_distribution = delayed_flights['Distance'].describe()
print("\nDistribution of flight distances for delayed flights:\n", distance_distribution)

plt.figure(figsize=(10, 6))
plt.hist(delayed_flights['Distance'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Distances for Delayed Flights')
plt.xlabel('Distance (miles)')
plt.ylabel('Number of Delayed Flights')
plt.axvline(average_distance_delayed, color='red', linestyle='dashed', linewidth=1, label='Average Distance')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

# How do delays differ by aircraft type?


```python
# Group by AircraftType and calculate average delay time
average_delay_by_aircraft = df.groupby('AircraftType')['DelayMinutes'].mean().sort_values(ascending=False)
print("\nAverage delay time by aircraft type:\n", average_delay_by_aircraft)

plt.figure(figsize=(12, 6))
average_delay_by_aircraft.plot(kind='bar', color='lightcoral')
plt.title('Average Delay Time by Aircraft Type')
plt.xlabel('Aircraft Type')
plt.ylabel('Average Delay (Minutes)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
